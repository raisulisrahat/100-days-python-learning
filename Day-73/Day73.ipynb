```python
import matplotlib.pyplot as plt
import numpy as np

plt.title("Man and Women Ages")
# Plot 1 Men
x = np.array([10,20,30,60,40])
y = np.array([11,5,30,48,20])
plt.xlabel('Man')
plt.subplot(1,2,1)
plt.plot(x,y)
plt.grid()


# Plot 2 Women
a = np.array([10,11,15,23,50])
b = np.array([15,22,18,41,65])
plt.xlabel('Women')
plt.subplot(1,2,2)
plt.plot(a,b)
plt.grid()


plt.show()
```
