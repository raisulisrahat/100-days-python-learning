```python
# Matplotlib Pyplot Demo

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 10])

plt.plot(x, y)
plt.show
```

```python
# Matplotlib Pyplot Marker

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6, 1, 25, 3, 50])

plt.plot(x, marker='o' ) 
plt.show
```
